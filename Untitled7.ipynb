```python


dataset_path = "/content/drive/MyDrive/MultiZoo/train/train"

# Dönüştürme işlemleri
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# Veri yükleme ve ayırma
full_dataset = datasets.ImageFolder(dataset_path, transform=train_transform)
train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])
val_dataset.dataset.transform = val_transform

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

print("✅ Eğitim örnekleri:", len(train_dataset))
print("✅ Doğrulama örnekleri:", len(val_dataset))
print("✅ Sınıflar:", full_dataset.classes)
```

```python
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import precision_score, recall_score, f1_score
import copy

# Cihaz ayarı
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Model kur
num_classes = len(train_loader.dataset.dataset.classes)
model = timm.create_model('vit_base_patch16_224', pretrained=True)
model.head = nn.Linear(model.head.in_features, num_classes)
model = model.to(device)

# Kayıp fonksiyonu ve optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# Eğitim fonksiyonu
def train(model, dataloader):
    model.train()
    loss_total, y_true, y_pred = 0, [], []
    for x, y in dataloader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        out = model(x)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()
        loss_total += loss.item()
        y_true.extend(y.cpu().numpy())
        y_pred.extend(out.argmax(1).cpu().numpy())
    acc = np.mean(np.array(y_true) == np.array(y_pred))
    return loss_total/len(dataloader), acc, \
        precision_score(y_true, y_pred, average="macro", zero_division=0), \
        recall_score(y_true, y_pred, average="macro", zero_division=0), \
        f1_score(y_true, y_pred, average="macro", zero_division=0)

# Doğrulama fonksiyonu
def validate(model, dataloader):
    model.eval()
    loss_total, y_true, y_pred = 0, [], []
    with torch.no_grad():
        for x, y in dataloader:
            x, y = x.to(device), y.to(device)
            out = model(x)
            loss = criterion(out, y)
            loss_total += loss.item()
            y_true.extend(y.cpu().numpy())
            y_pred.extend(out.argmax(1).cpu().numpy())
    acc = np.mean(np.array(y_true) == np.array(y_pred))
    return loss_total/len(dataloader), acc, \
        precision_score(y_true, y_pred, average="macro", zero_division=0), \
        recall_score(y_true, y_pred, average="macro", zero_division=0), \
        f1_score(y_true, y_pred, average="macro", zero_division=0)
```

```python
from torch.optim import AdamW
from torch.optim.lr_scheduler import ReduceLROnPlateau
import copy
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Hiperparametreler
num_epochs = 50
patience = 5
counter = 0
best_acc = 0

# Performans takip listeleri
train_losses, val_losses = [], []
train_accs, val_accs = [], []
train_f1s, val_f1s = [], []

# Optimizer ve LR scheduler
optimizer = AdamW(model.parameters(), lr=1e-4)
scheduler = ReduceLROnPlateau(
    optimizer, mode='max', factor=0.5, patience=3, verbose=True
)

# Eğitim döngüsü
for epoch in range(num_epochs):
    print(f"\n📘 Epoch {epoch+1}/{num_epochs}")

    tr_loss, tr_acc, tr_prec, tr_rec, tr_f1 = train(model, train_loader)
    val_loss, val_acc, val_prec, val_rec, val_f1 = validate(model, val_loader)

    # Listeye ekle
    train_losses.append(tr_loss)
    val_losses.append(val_loss)
    train_accs.append(tr_acc)
    val_accs.append(val_acc)
    train_f1s.append(tr_f1)
    val_f1s.append(val_f1)

    # Metrikleri yazdır
    print(f"✅ Train  Loss: {tr_loss:.4f} | Acc: {tr_acc:.4f} | Prec: {tr_prec:.4f} | Rec: {tr_rec:.4f} | F1: {tr_f1:.4f}")
    print(f"🔎 Valid  Loss: {val_loss:.4f} | Acc: {val_acc:.4f} | Prec: {val_prec:.4f} | Rec: {val_rec:.4f} | F1: {val_f1:.4f}")

    # En iyi modeli kaydet
    if val_acc > best_acc:
        best_acc = val_acc
        best_weights = copy.deepcopy(model.state_dict())
        torch.save(best_weights, "best_model.pth")
        print("💾 Yeni en iyi model kaydedildi.")
        counter = 0
    else:
        counter += 1
        print(f"⚠️ Gelişme yok. EarlyStopping sayacı: {counter}/{patience}")
        if counter >= patience:
            print("⛔ EarlyStopping tetiklendi.")
            break

    # LR scheduler çalıştır
    scheduler.step(val_acc)

# 📈 Accuracy Eğrisi
plt.figure(figsize=(10,5))
plt.plot(train_accs, label="Train Accuracy", color='orange')
plt.plot(val_accs, label="Validation Accuracy", color='green')
plt.title("Accuracy Learning Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# 📈 F1-Score Eğrisi
plt.figure(figsize=(10,5))
plt.plot(train_f1s, label="Train F1 Score", color='purple')
plt.plot(val_f1s, label="Validation F1 Score", color='red')
plt.title("F1 Score Learning Curve")
plt.xlabel("Epoch")
plt.ylabel("F1 Score")
plt.legend()
plt.grid(True)
plt.show()

# 📊 Metrikleri CSV'ye yaz
df_metrics = pd.DataFrame({
    "Epoch": list(range(1, len(train_accs)+1)),
    "Train Accuracy": train_accs,
    "Validation Accuracy": val_accs,
    "Train F1 Score": train_f1s,
    "Validation F1 Score": val_f1s
})
df_metrics.to_csv("training_metrics.csv", index=False)

# 💾 CSV dosyasını indir
from google.colab import files
files.download("training_metrics.csv")
```
